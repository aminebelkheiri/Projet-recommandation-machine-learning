```python
# Fonction permettant de retourner un graphique de frquence d'apparitions des acteurs

def frequenceActeurs():
    final=pd.concat([data.actor_1_name, data.actor_2_name,data.actor_3_name])
    acteurs = final.value_counts().rename_axis('acteurs').to_frame('apparitions')
    acteurs1 = acteurs [0:10]
    
    fig, ax1 = plt.subplots(figsize=(17, 10))
    sns.barplot(acteurs1.index, acteurs1.apparitions)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=16)
    ax1.set_ylabel("apparitions",fontsize=14)
    ax1.set_xlabel("",fontsize=14)
    plt.title("Frquence d'apparitions des acteurs",fontsize=20)

    return plt.show()

frequenceActeurs()
```

```python
# Fonction permettant de gnrer un graphique sur les genres des films

def frequenceGenre():    
    
    dataGenre = data.genres.str.split("|",expand=True)

    genre1 = dataGenre[0]
    genre2 = dataGenre[1]
    genre3 = dataGenre[2]
    genre4 = dataGenre[3]
    genre5 = dataGenre[4]
    genre6 = dataGenre[5]
    genre7 = dataGenre[6]
    genre8 = dataGenre[7]


    finalgenre=pd.concat([genre1, genre2,genre3,genre4,genre5,genre6,genre7,genre8])
    
    genre = finalgenre.value_counts().rename_axis('genre').to_frame('frequence')
    genre1 = genre [0:15]
    
    fig, ax1 = plt.subplots(figsize=(17, 10))
    sns.barplot(genre1.index, genre1.frequence)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    ax1.set_ylabel("frquence",fontsize=14)
    ax1.set_xlabel("",fontsize=14)
    plt.title("Genres",fontsize=20)

    return plt.show()

frequenceGenre()
```

```python
def minus_sanEspace(x):
    if isinstance(x, list):
        return [str.lower(i.replace(" ", "")) for i in x]
    else:       
        if isinstance(x, str):
            return str.lower(x.replace(" ", ""))
        else:
            return ''
```

```python
# Voici les colonnes qui ncessitent l'application de cette fonction
les4colonnes = ["director_name", "plot_keywords", "genres", "the_actors"]

for element in les4colonnes:
    madata[element] = madata[element].apply(minus_sanEspace)
```

```python
madata['spaggheti'] = madata.apply(long_spaggheti, axis=1)
```

## Construction de l'algorithme de recommandation

```python
# Importons CountVectorizer et crons le count matrix
from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(madata['spaggheti'])
```

```python
# Calculons la matrice de similarit des cosinus en fonction de count_matrix
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim2 = cosine_similarity(count_matrix, count_matrix)
```

```python
# Rinitialisons l'index de notre DataFrame principal et construisons le reverse mapping comme avant
madata = madata.reset_index()
indices = pd.Series(madata.index, index=madata['movie_title']).drop_duplicates()
```

```python
# Voici la fonction qui prend un film en entre et ressort les films les plus similaires
def movies_recommendations(title, cosine_sim=cosine_sim2):
    # Obtenir l'index du film qui correspond au titre
    idx = indices[title]

    # Obtenez les scores de similarit pairwsie de tous les films avec ce film
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Trier les films en fonction des scores de similarit
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Obtenez les scores des 5 films les plus similaires
    sim_scores = sim_scores[1:6]

    # Obtenir les index du film
    movie_indices = [i[0] for i in sim_scores]

    # Retourne le top 5 des films les plus similaires
    return madata['movie_title'].iloc[movie_indices]
```
